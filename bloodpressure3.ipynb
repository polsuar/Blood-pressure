```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
%matplotlib inline
```

```python
raisedpb_df = pd.read_csv('https://raw.githubusercontent.com/polsuar/Blood-pressure/main/raisedpb.csv', skiprows=[1,2,3,4])
raisedpbglobal_df = pd.read_csv("https://raw.githubusercontent.com/polsuar/Blood-pressure/main/raisedpbglobal.csv", skiprows=[1,2,3])
```

```python
raisedpb_df.head()
```

```python
raisedpbglobal_df.head()
```

```python
cols = raisedpb_df.keys()
print(cols)
```

```python
#El método DataFrame.loc [] es un método de Pandas que solo toma etiquetas de índice y devuelve una fila o un dataframe(en este caso un df)
raisedpb = raisedpb_df.loc[:, cols[1]:cols[-1]]
```

```python
dates = raisedpb.keys()
dates=dates.tolist()
dates.reverse()
```

```python
cases_argentina = []
cases_us = []
cases_global = []
```

```python
for i in dates:
  cases_argentina.append(raisedpb_df[raisedpb_df['Unnamed: 0']=='Argentina'][i].sum())
  cases_us.append(raisedpb_df[raisedpb_df['Unnamed: 0']=='United States of America'][i].sum())

for i in dates:
  cases_global.append(raisedpbglobal_df[raisedpbglobal_df['Unnamed: 0']=='Global'][i].sum())
  
```

```python
for i in range(0, len(dates)): 
  dates[i] = int(dates[i]) 
  
```

```python
num_us=[]
for i in cases_us:
  a = i.split()
  num_us.append(float(a[0]))

num_arg=[]
for i in cases_argentina:
  a = i.split()
  num_arg.append(float(a[0]))

num_global=[]
for i in cases_global:
  a = i.split()
  num_global.append(float(a[0]))
```

```python
plt.figure(12)
plt.figure(figsize=(12,6))
plt.plot(dates, num_global, c="g", label="Global")
plt.plot(dates, num_arg, c="b", label="Argentina")
plt.plot(dates, num_us, c="r", label="US")
plt.legend(frameon=False, loc=0)
plt.xticks(rotation=45,fontsize=10)
ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)    
ax.spines["right"].set_visible(False)    
plt.ylim(0, 50)    
plt.xlim(1975, 2015)
for y in range(10, 50, 10):    
    plt.plot(range(1975, 2017), [y] * len(range(1975, 2017)), "--", lw=0.5, color="black", alpha=0.3) 
plt.xlabel("Year",fontsize=14)
plt.ylabel("Prevalence of raised blood pressure(%)",fontsize=14)
plt.title("Percent of 18+ population with raised blood pressure (1975-2015)",fontsize=17)
plt.text(1973, -12,"Data source: https://apps.who.int/gho/data/view.main.NCDBPAREGv?lang=en \nhttps://apps.who.int/gho/data/view.main.2464ESTSTANDARD?lang=en"
 "\nGlobal Health Observatory data repository - WHO", fontsize=9)
```
